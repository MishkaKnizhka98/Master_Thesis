```python
import matplotlib.pyplot as plt
import numpy as np
import mplcursors
```

```python
# Function to plot
def curve_function(x):
    return np.sin(x)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Generate x values
x_values = np.linspace(0, 10, 100)
y_values = curve_function(x_values)

# Plot the initial curve
line, = ax.plot(x_values, y_values, label='Curve')
```

```python
# Enable draggable points with mplcursors
mplcursors.cursor(hover=True)
```

```python
# Function to update the curve when dragged
def update_curve(index, x, y):
    y_values = curve_function(x_values + x)
    line.set_ydata(y_values)
    fig.canvas.draw_idle()
```

```python
# Connect the update_curve function to the cursor
mplcursors.connect("add", update_curve)
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

# Function to plot
def curve_function(x, shift):
    return np.sin(x + shift)

# Create a figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

# Initial shift value
initial_shift = 0.0

# Generate x values
x_values = np.linspace(0, 10, 100)
y_values = curve_function(x_values, initial_shift)

# Plot the initial curve
line, = ax.plot(x_values, y_values, label='Curve')

# Add sliders for adjusting parameters
axcolor = 'lightgoldenrodyellow'
ax_shift = plt.axes([1, 0.1, 0.65, 0.03], facecolor=axcolor)
s_shift = Slider(ax_shift, 'Shift', -5.0, 5.0, valinit=initial_shift)

# Function to update the curve when sliders are changed
def update(val):
    shift = s_shift.val
    y_values = curve_function(x_values, shift)
    line.set_ydata(y_values)
    fig.canvas.draw_idle()

# Attach the update function to the sliders
s_shift.on_changed(update)

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

plt.show()
```
